```python
import networkx as nx
import gzip
from collections import defaultdict
import numpy as np
```

```python
graph_path = "/Users/matthew/Documents/Efficient Gaussian Process on Graphs/Efficient_Gaussian_Process_On_Graphs/experiments_sparse/youtube/data/com-youtube.ungraph.txt.gz"
top_5000_path = "/Users/matthew/Documents/Efficient Gaussian Process on Graphs/Efficient_Gaussian_Process_On_Graphs/experiments_sparse/youtube/data/com-youtube.top5000.cmty.txt.gz"
```

```python
G = nx.Graph()

with gzip.open(graph_path, 'rt') as f:
    for line in f:
        if line.startswith('#'):
            continue
        u, v = map(int, line.strip().split())
        G.add_edge(u, v)

print(f"Loaded graph with {G.number_of_nodes()} nodes and {G.number_of_edges()} edges")
```

```python
num_nodes = G.number_of_nodes()
A = nx.adjacency_matrix(G)
```

```python
x = np.random.rand(num_nodes)
```

```python
y = A@x
```

```python
y.max(), y.min(), y.mean(), y.std()
```

```python
node_to_labels = defaultdict(set)

with gzip.open(top_5000_path, 'rt') as f:
    for community_id, line in enumerate(f):
        members = map(int, line.strip().split())
        for node in members:
            node_to_labels[node].add(community_id)

print(f"Loaded {len(node_to_labels)} labeled nodes")
```
